```python
import math
import random
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

def shor_classical(N):
    if N % 2 == 0: #check if N is even, if it is, 2 is a factor
        print("N is even")
        return 2, N // 2

    while True:
        x = random.randint(2, N - 1) # generate the random guess
        print("Trying x =", x)

        g = math.gcd(x, N) # compute gcd of x and N
        print("gcd(x, N) =", g)

        if g != 1: # if gcd is not 1, we found a factor
            print("Found factor by gcd")
            return g, N // g

        r = quantum_order(x, N) # need to change to quantum order finding
        print("Order r =", r)

        if r % 2 != 0: # shors algorithm only works for even powers
            print("r is odd, retrying")
            continue

        if pow(x, r // 2, N) == N - 1: # check if one of the factor is just a multiple of N
            print("Bad r, retrying")
            continue
        # find the two factors
        p = math.gcd(pow(x, r // 2) - 1, N) 
        q = math.gcd(pow(x, r // 2) + 1, N)
        # check if we found the factors
        if p * q == N:
            print("Factors of N:", p, q)
            return p, q


def classical_order(x, N): # function to find the order of x mod N
    r = 1
    value = x % N
    while value != 1:
        value = (value * x) % N
        r += 1
    return r

def quantum_order(x, N): # placeholder for quantum order finding
    n = math.ceil(math.log2(N))   # work qubits
    a = 2 * n                     # counting qubits

    count = QuantumRegister(a, "count")
    work  = QuantumRegister(n, "work")

    qc = QuantumCircuit(count, work)

    # STEP 1: Create superposition over |0> ... |2^a - 1>
    qc.h(count)

    # STEP 2: Initialize work register to |1>
    qc.x(work[0])
    qc.draw("text")
    return qc
    
print(shor_classical(15))
```
